```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
def visualize_data(paths, sample=25):
    n_cols = 5
    n_rows = 5
    fig = plt.figure(figsize=(10,10))
    for i in range(1, n_cols*n_rows+1):
        try:
            im = Image.open(paths[i + np.random.randint(1, len(paths)) ])
        except:
            continue
        plt.subplot(n_rows, n_cols, i)
        plt.imshow(im)
        plt.axis('off')
        
plt.show()
```

```python
paths = glob.glob(dir_path + "*")
visualize_data(paths)
```
